```python
import json
first_names = nd.get_top_names(n=1000, country_alpha2='US', use_first_names=True)

with open('./data/top_first_names.jsonl', 'w') as file:
    for gender, names in first_names['US'].items():
        for name in names:
            # Create a dictionary for each name
            entry = {
                "first_name": name,
                "gender": "Male" if gender == 'M' else "Female",
                "country_alpha2": "US"
            }
            # Write the JSON string followed by a newline
            file.write(json.dumps(entry) + '\n')
```

```python
import json
last_names = nd.get_top_names(n=10000, country_alpha2='US', use_first_names=False)

with open('./data/top_last_names.jsonl', 'w') as file:
    for names in last_names['US']:
        # Create a dictionary for each name
        entry = {
            "last_name": names,
            "country_alpha2": "US"
        }
        # Write the JSON string followed by a newline
        file.write(json.dumps(entry) + '\n')
```

```python
import jsonlines

first_names = []
with jsonlines.open('./data/top_first_names.jsonl') as reader:
    for line in reader:
        first_names.append(line['first_name'])
print(first_names)

last_names = []
with jsonlines.open('./data/top_last_names.jsonl') as reader:
    for line in reader:
        last_names.append(line['last_name'])
print(last_names)
```
